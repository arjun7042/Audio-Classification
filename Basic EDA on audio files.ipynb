```python
import matplotlib.pyplot as plt
import IPython.display as ipd
import librosa
import librosa.display
import seaborn as sns
```

## Data Visualization

```python
filename = "UrbanSound8k/dog_bark.wav"
```

```python
### Dog Sound
plt.figure(figsize=(14,5))
data,sample_rate=librosa.load(filename)
librosa.display.waveshow(data,sr=sample_rate)
ipd.Audio(filename)
```

```python
filename1 = "UrbanSound8k/gun_shot.wav"
```

```python
### Gun_shot Sound
plt.figure(figsize=(14,5))
data,sample_rate=librosa.load(filename1)
librosa.display.waveshow(data,sr=sample_rate)
ipd.Audio(filename1)
```

```python
sample_rate
```

Librosa always keep sample rate of any audio file to 22KHZ

```python
data
```

We can see that audio data has been normalized between -1 and 1

## Metadata of all sound files

```python
import pandas as pd

metadata=pd.read_csv('UrbanSound8K/metadata/UrbanSound8K.csv')
metadata.head(10)
```

```python
### Check whether the dataset is imbalanced
metadata['class'].value_counts()
```

```python
plt.figure(figsize=(15,6))
sns.countplot(x ='class', data = metadata, palette = "Set2")
plt.show()
```

Its not a imbalanced dataset as most of the categories are having good amount of datapoints
